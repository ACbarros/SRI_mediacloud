Recuperação de Informação de HTML bruto de notícias genéricas
====================
Fundamentos
-----------
Como o código HTML é gerado automaticamente pela maior parte dos editores de website, podemos assumir a ocorrência de padrões estruturais que facilitam a recuperação de informação. Dentre eles:

- O texto principal não está contido em tags HTML ou dentro de scripts.

- O texto principal ocupa uma porção 'razoávelmente' contígua do documento. Isto é, raros blocos de propaganda, menus ou imagens entre parágrafos.

- Parágrafos textuais costumam ser formados por sentenças de mais de 5 palavras


Estratégia
-------------
A análise de cada texto pode acabar sendo extremamente custosa, a ponto de exigir na ordem de segundos por texto. Ela consiste nas seguintes etapas:

- Remoção de Scripts: varremos o documento e retiramos todas as regiões de script, composto recursivamente.

- Remoção de Tags HTML: varremos o documento e retiramos todas as regiões de código HTML, realizado iterativamente.

- Alinhamento: Para analizar os documentos por comprimento das linhas, desfazemos a identação do HTML e esvaziamos as linhas com menos que 5 palavras.

- Buscamos a região de pico no comprimento das linhas e definimos uma vizinhança onde o comprimento da linha é semelhante.


Observações
-------------
- Não se pode confiar que haverá apenas uma seção &lt;body>&lt;/body> ... Exemplo: http://www12.senado.gov.br/noticias/materias/2013/12/12/comissao-aprova-nova-lei-de-licitacoes

```python
# Get collection of html documents from the dicrectory raw_data/
import glob
raw_data = glob.glob("raw_data/*")

print raw_data[:3]

raw_docs = list()
for path in raw_data:
    f = open(path,'r')
    raw_docs.append(f.read())
    f.close()
```

```python
def shortStatement(string):
    if(len(string.split(" ")))<5:
        return True
    else:
        return False

def unIdent(text):
    after = len(text)
    before = after+1
    while( after<before ):
        before = after
        text   = text.replace("  "," ")
        after  = len(text)
    text   = text.replace("\n ","\n")
    return text

def cleanScripts(html):
    start = html.find("<script")
    end = html.find("</script>")
    if start==-1:
        return html
    return html[:start]+cleanScripts(html[end+9:])

def cleanTags(html):
    listed = list(html)
    i = 0
    eraser = False
    while(i<len(listed)):
        if(listed[i]=='<'):
            eraser=True
        while(eraser):
            if(listed[i]=='>'):
                eraser=False
                del(listed[i])
                i -= 1
            else:
                del(listed[i])
        i += 1
    return "".join(listed)

def cleanHTML(html):
    clean = unIdent(cleanTags(cleanScripts(html)).replace("	",""))
    linelist = clean.split('\n')
    i = 0
    while(i<len(linelist)):
        if shortStatement(linelist[i]) :
            linelist[i]='\n'
        i += 1
    return "\n".join(linelist)

%timeit clean_docs = [cleanHTML(html) for html in raw_docs[:50] ]
```

```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt


for doc in clean_docs[:10]:
    linelenghts = [len(line) for line in doc.split('\n')]
    lenframe = pandas.DataFrame(linelenghts, columns=['Size'])
    lenframe.plot()
```
