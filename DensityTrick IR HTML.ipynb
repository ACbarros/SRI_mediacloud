Recuperação de Informação de HTML bruto de sites de notícias
====================
Fundamentos
-----------
Como o código HTML é gerado automaticamente pela maior parte dos editores de website, podemos assumir a ocorrência de padrões estruturais que facilitam a recuperação de informação. Dentre eles:

- O texto principal não está contido em tags HTML, scripts ou campos de formatação

- O texto principal ocupa uma porção 'razoávelmente' contígua do documento. Isto é, raros blocos de propaganda, menus ou imagens entre parágrafos.

- Parágrafos textuais costumam ser formados por sentenças de mais de 5 palavras, gerando linhas longas


Estratégia
-------------
Limpar o documento:

- Esvaziar o documento de scripts e styles (sem remover as linhas)

- Remover tags &lt;p> e &lt;div>&lt;/div>, que frequentemente são usadas no interior do texto para demarcar parágrafos. (Protege os textos por causa da etapa seguinte)

- Transformar todas as tags de HTML em '\n', tornando o documento mais esparso exceto na região densa em texto, que (torçamos!) contém típicamente menos código HTML.

Destacar áreas relevantes:

- Localizar regiões de pico na quantidade de caracteres por linha, caracterizando concentração de texto.

- Não basta ser uma linha grande, tem que estar próximo a outras linhas grandes (no caso de um documento identado nos parágrafos) ou ser muito maior que o resto (no caso de parágrafos aglomerado em poucas linhas). Portanto para detectar picos usamos a moving average ao longo do documento.

Precisa melhorar
---------------

- Mecanismo de decisão se um pico é relevante ou não. 
- - Atualmente: mavgA - mavgB > 10
- - Proposta: partir do máximo global até encontrar mínimos globais (usa o fato do texto ser contínuo)

Observações
-------------
- Não se pode confiar que haverá apenas uma seção &lt;body>&lt;/body> ... Exemplo: http://www12.senado.gov.br/noticias/materias/2013/12/12/comissao-aprova-nova-lei-de-licitacoes

```python
# Obtendo HTML de sites para testes
import glob
raw_data = glob.glob("raw_data/*")

print "{} Documentos no banco: ".format(len(raw_data)),raw_data[:3]

raw_docs = list()
for path in raw_data:
    f = open(path,'r')
    raw_docs.append(f.read())
    f.close()
```

Limpando documentos HTML
-------------

```python
# Remove a identação de um texto
def unIdent(text):
    text = text.replace("	","")
    after = len(text)
    before = after+1
    while( after<before ):
        before = after
        text   = text.replace("  "," ")
        after  = len(text)
    text   = text.replace("\n ","\n")
    return text

# Remove todas as tags HTML de um documento html
def cleanTags(html):
    listed = list(html)
    i = 0; eraser = False
    while(i<len(listed)):
        if(listed[i]=='<'): # entra no modo deletar
            eraser=True
        while(eraser):
            if(listed[i]=='>'): # sai do modo deletar se encontrar fim de tag
                eraser=False
                listed[i]='\n'
                #del(listed[i])
                #i -= 1
            elif(listed[i]!='\n'): # não apagar linefeeds para manter bijeção de linhas
                del(listed[i])
            else:
                i += 1
        i += 1
    return "".join(listed)

# Remove todos os scripts (entre tags <script></script>), fazer antes de cleanTags
def cleanScripts(html):
    start = html.find("<script")
    end = html.find("</script>")
    while(start!=-1):
        listed = list(html)
        for i in range(start,end+9):
            if(listed[i]!='\n'):
                listed[i]=" "
        html = "".join(listed)
        start = html.find("<script")
        end = html.find("</script>")
    return html

# Remove a seção <style></style>, fazer antes de cleanTags
def cleanStyle(html):
    start = html.find("<style")
    end = html.find("</style>")
    while(start!=-1):
        listed = list(html)
        for i in range(start,end+8):
            if(listed[i]!='\n'):
                listed[i]=" "
        html = "".join(listed)
        start = html.find("<style")
        end = html.find("</style>")
    return html

def cleanHTML(html):
    html = html.replace("<p>"," ").replace("<div>", " ").replace("</div>", " ")
    noscripts = cleanScripts(html)
    nostyle = cleanStyle(noscripts)
    notags = cleanTags(nostyle)
    txt = unIdent(notags)
    return txt
```

Decidir área relevante
------------------

```python
periodA = 30
periodB = 100

def relevanceGraph(html,label="Value"):
    clean = cleanHTML(html).split('\n')
    lenghts = [len(line) for line in clean]
    lenf = pandas.DataFrame(lenghts, columns=["Lenghts " + label])

    mavgA = pandas.rolling_mean(lenf,periodA).shift(-periodA/2)
    mavgA.columns=["mavgA " + label]
    mavgB = pandas.rolling_mean(lenf,periodB).shift(-periodB/2)
    mavgB.columns=["mavgB " + label]
    fig = mavgA.plot()
    mavgB.plot(ax=fig)
    
    (lenf / 10).plot()
    spikeness = mavgA["mavgA " + label] - mavgB["mavgB " + label]
    spikeness.plot(label="Spikeness")
    
    spikeness = spikeness.to_dict()
    threshold = 10
    for l in spikeness.keys():
        if spikeness[l]>threshold:
            print clean[l]
```
